## 4.グラフの色・凡例・軸の編集<a name="凡例等"></a>

```python
#sample
import numpy as np
import pandas as pd


data = pd.DataFrame()
np.random.seed(1)
data["A"] = (
    list(np.random.normal(10, 10, 25))
    + list(np.random.normal(18, 10, 25))
    + list(np.random.normal(25, 10, 25))
    + list(np.random.normal(32, 10, 25))
)
data["B"] = ["A"] * 25 + ["B"] * 25 + ["C"] * 25 + ["D"] * 25
data["C"] = list(np.random.normal(100, 10, 50)) + list(np.random.normal(10, 3, 50))
data["D"] = np.random.normal(10, 4, 100)
```

## 4-A.グラフの色<a name="グラフの色"></a>
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html?highlight=plot#matplotlib.axes.Axes.plot

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#列Bの値で色分け
for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values)
plt.show()
```

## 4-B.凡例<a name="凡例"></a>
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.legend.html

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 凡例を右上に作成
for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values, label=value_b)

ax.legend(loc="upper right")

plt.show()
```

## 4-B.ヒストグラムの色分け・凡例<a name="凡例"></a>

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.hist(data.loc[data["B"] == value_b, "D"].values, label=value_b)

ax.legend()

plt.show()
```

## 4-B.不透明・ビンの数を増やす<a name="凡例"></a>
https://matplotlib.org/api/_as_gen/matplotlib.patches.Patch.html#matplotlib.patches.Patch.set_alpha
- bins : 棒の数

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.hist(
        data.loc[data["B"] == value_b, "D"].values, label=value_b, alpha=0.5, bins=10
    )

ax.legend()

plt.show()
```

## 4-B.y軸を対数スケール<a name="凡例"></a>
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_yscale.html?highlight=set_yscale#matplotlib.axes.Axes.set_yscale

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.hist(
        data.loc[data["B"] == value_b, "D"].values, label=value_b, alpha=0.5, bins=10
    )

ax.set_yscale("log")
ax.legend()

plt.show()
```

## 4-B.軸ラベルのサイズ<a name="凡例"></a>
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.tick_params.html?highlight=tick_params#matplotlib.axes.Axes.tick_params

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values, label=value_b)
ax.legend(loc="upper right")

ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)

plt.show()
```

## 4-B.凡例をグラフ本体の右側に配置<a name="凡例"></a>

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values, label=value_b)

ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1))

ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)

plt.show()
```

## 4-B.凡例のタイトルをフォントサイズが20の'列Bの値'<a name="凡例"></a>
https://matplotlib.org/gallery/api/font_family_rc_sgskip.html

```python
plt.rcParams['font.family'] = 'Yu Mincho'

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values, label=value_b)

lg = ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1))
lg.set_title("列Bの値", prop={"size": 20})

ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)

plt.show()
```

## 4-B.X', 'Y'と軸ラベルを設定<a name="凡例"></a>
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html?highlight=ax%20set_xlabel#matplotlib.axes.Axes.set_xlabel
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html?highlight=ax%20set_ylabel#matplotlib.axes.Axes.set_ylabel

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values, label=value_b)
    
lg = ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1))
lg.set_title("列Bの値", prop={"size": 20})

ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)

ax.set_xlabel("X", fontsize=18)
ax.set_ylabel("Y", fontsize=18)

plt.show()
```

## 4-B.y軸の範囲の変更<a name="凡例"></a>
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_ylim.html?highlight=ax%20set_ylim#matplotlib.axes.Axes.set_ylim

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for value_b in data["B"].unique():
    ax.plot(data.loc[data["B"] == value_b, "A"].values, label=value_b)
lg = ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1))
lg.set_title("列Bの値", prop={"size": 20})
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
ax.set_xlabel("X", fontsize=18)
ax.set_ylabel("Y", fontsize=18)

ax.set_ylim(0, 50)


plt.show()
```
